```python
from PIL import Image
import os
# load_images and captions, tokens indica que el texto de la dataset seran los tokens, de lo contrario sera un caption comun.
def load_data(base_path = '/content/gdrive/My Drive/Images-RI-ML/Data/', tokens = False):
    data = [file for file in os.listdir(base_path) if file.endswith(('.jpg', '.png'))]

    images  = []
    captions  = []

    for load_image_path in data:
        image = Image.open(base_path + load_image_path)
        images.append(image)
        text_path = base_path + load_image_path[:-4]+'.txt'
        if not tokens:
            text_path = base_path + load_image_path[:-4]+'_caption.txt'

        with open(text_path) as file:
            text = file.read()
            file.close()
            captions.append(text)
    return {'image': images, 'caption': captions}
```

```python
from datasets import Dataset
dataset = Dataset.from_dict(load_data())
```

```python
import matplotlib.pyplot as plt
LEN = 10
INDEX = 0
for image, text in zip(dataset['image'],dataset['caption']):
    INDEX+=1
    if INDEX > LEN: break
    plt.figure(figsize=(2,2))
    plt.title(text)
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```
